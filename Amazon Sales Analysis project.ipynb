# Impoer libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Amazon.csv")
```

```python
df
```

# Data Cleaning

```python
df.columns
```

```python
df.drop(columns=["New" , "PendingS"] , axis=1 , inplace=True)
```

```python
df.columns
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df["currency"] = df["currency"].bfill()
df["Amount"] = df["Amount"].fillna(df["Amount"].mean())

```

```python
df.isnull().sum()
```

```python
mode_value = df["ship-city"].mode()[0]
df["ship-city"] = df["ship-city"].fillna(mode_value)


mode_value = df["ship-state"].mode()[0]
df["ship-state"] = df["ship-state"].fillna(mode_value)



mode_value = df["ship-country"].mode()[0]
df["ship-country"] = df["ship-country"].fillna(mode_value)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['ship-postal-code'], inplace=True)
```

```python
df["fulfilled-by"] = df["fulfilled-by"].bfill()
```

```python
df["fulfilled-by"].unique()
```

```python
# Replace 'nan' strings with NaN values
df["fulfilled-by"].replace("nan", np.nan, inplace=True)

# Drop rows where 'fulfilled-by' column has NaN values
df.dropna(subset=["fulfilled-by"], inplace=True)
```

```python
df.info()
```

```python
df['ship-postal-code'] = df['ship-postal-code'].astype('int64')
```

```python
df.info()
```

```python
df
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.rename(columns={"Qty":"Quantity"} , inplace=True)
```

```python
df
```

```python
# Remove all rows of quantity column where value is 0
df = df[df['Quantity'] != 0]
```

```python
df_zero_quantity= df.loc[df['Quantity'] == 0]
df_zero_quantity
```

```python
df.describe()
```

```python
df.describe(include="object")
```

# Exploratory Data Analysis:

# Category

```python
 df["Category"].value_counts()
```

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(x="Category" , data=df , palette="Dark2" , hue="Category")

for bars in ax.containers:
    ax.bar_label(bars)
```

Note: From the above graph it is clear most of the sale items are T-shirt and shirt

Note: Now find which size category most sold out

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(x="Size" , data=df , palette="Dark2" , hue="Size")

for bars in ax.containers:
    ax.bar_label(bars)
```

Note: Most of the people buy Medium size items

```python
plt.figure(figsize=(10,5))
sales_gen=df.groupby(["Category"], as_index=False)["Quantity"].sum().sort_values(by="Quantity" , ascending=False)
ax=sns.barplot(x="Category" , y="Quantity" , data=sales_gen , palette="Dark2" , hue="Category")
plt.title("Quantity of each Cateogry")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

# Amount per Category

```python
plt.figure(figsize=(10,5))
sales_gen=df.groupby(["Category"], as_index=False)["Amount"].sum().sort_values(by="Amount" , ascending=False)
ax=sns.barplot(x="Category" , y="Amount" , data=sales_gen , palette="Dark2" , hue="Category")
plt.title("Amount per Cateogry")
plt.ylim(0, sales_gen['Amount'].max() * 1.1)  # Adjust the multiplier for padding if needed
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

#It show T-shirt generate more revenue

# Time

```python
df['Month'] = df['Date'].dt.month
```

```python
df['Month'].unique()
```

```python
import calendar
```

```python
df["month_names"] = [calendar.month_name[month] for month in df['Month']]
```

```python
df["month_names"].unique()
```

# Find out how many items sales out and create how much in each month

```python
# Plotting with seaborn countplot
plt.figure(figsize=(15, 8))  # Adjust figure size if needed
ax=sns.countplot(x='month_names', hue='Category', data=df, palette='Set1')

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Count of Each Category in Each Month')

# Adjusting legend
plt.legend(title='Category')
for bars in ax.containers:
    ax.bar_label(bars)
# Adjusting legend
# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10, 8))  # Adjust figure size if needed
ax=sns.barplot(x='month_names', y='Amount', hue='Category', data=df, palette='Set1' ,  estimator="sum")

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Amount Generated by Category and Month')

# Adjusting legend
plt.legend(title='Category')
plt.show()
```

Above graph show which Category items create how much money or how much sold out in which month

# State and city

```python
plt.figure(figsize=(12,6))
sns.countplot(x="ship-state" , data=df , palette="Dark2" , hue="ship-state")
plt.xlabel("ship-state")
plt.ylabel("count")
plt.xticks(rotation=90)
plt.show()
```

```python
state_counts = df['ship-state'].value_counts().reset_index()
state_counts.columns = ['ship-state', 'count']
top_10_states = state_counts.nlargest(10, 'count')
top_10_states
```

```python
plt.figure(figsize=(10, 6))
ax=sns.barplot(x='ship-state', y='count', data=top_10_states, palette='viridis' , hue="ship-state")
plt.xlabel('state')
plt.ylabel('Count')
plt.title('Top 10 state by Count')
plt.xticks(rotation=45)
plt.tight_layout()
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
city_counts = df['ship-city'].value_counts().reset_index()
city_counts.columns = ['ship-city', 'count']
top_10_cities = city_counts.nlargest(10, 'count')
top_10_cities
```

```python
plt.figure(figsize=(10, 6))
ax=sns.barplot(x='ship-city', y='count', data=top_10_cities, palette='viridis' , hue="ship-city")
plt.xlabel('city')
plt.ylabel('Count')
plt.title('Top 10 cities by Count')
plt.xticks(rotation=45)
plt.tight_layout()
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```
